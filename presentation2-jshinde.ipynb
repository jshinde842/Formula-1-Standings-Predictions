## Part 2: Linear Regression Analysis

Let's analyze the linear models we derived from part 1.

Recall the two lines of best fit derived from Linear Regression over Lewis Hamilton and Valtteri Bottas' race data.



```julia
using Plots
```

#### Lewis Hamilton

```julia
l(x) = -0.03019339*x + 5.79165713
numRacesToExpectedFirst = (5.79165713-1)/0.03019339
print("y = 1 intercept: ")
print(numRacesToExpectedFirst)
plot(l, 0, 160, xaxis = "Races", yaxis = "Finishing Position")
```

From the Linear Regression model derived in part 1, we can see the consistency of Lewis Hamilton's improvement throughout his career. The equation let's us come to the conclusion that Lewis Hamilton would be expected to finish first in every grand prix in the future after 158 races.

#### Valtteri Bottas

```julia
v(x) = -0.06664188*x + 12.02118485
numRacesToExpectedFirst = (12.02118485-1)/0.06664188
print("y = 1 intercept: ")
print(numRacesToExpectedFirst)
plot(v, 0, 170, xaxis = "Races", yaxis = "Finishing Position")
```

Valtteri Bottas' data is interesting as well. He shows the some consistency in improvement. Valtteri is expected to consistently finish first in every grand prix after 165 races.

### Bias

You may have noticed that according to both Lewis Hamilton and Valtteri Bottas' performances they would both be projected to win first place after a certain amount of races. However, there are many factors which induce bias and can have a serious effect on the projections we calculated.

Here are some factors. First is team budgets.

<img src="budget.png" alt="Drawing" style="width: 500px;"/>

Not only do teams that have more budget tend to do better, but also if a team starts struggling financially there would be serious fallbacks on the drivers and their performances. If Mercedes were to suffer from some financial issues, we'd see a skew in the data and would turn the regression into something much less accurate. It would be better to model this scenario parabolically instead.

Another factor which was mentioned in part 1 is the role of the driver in the team. Valtteri Bottas would have a harder time consistently finishing 1st when he is considered a support driver to Lewis Hamilton. 

TO Conclude:

We can use mathematical models to model virtually anything, however, in the real world there are always unknown elements and factors which adversely affect the models you create. This is due to bias which can be purposeful or not. That's why I wanted to show you the race data between two remarkable F1 drivers and show that although they can be expected to win more, the bias imposed by external factors will always propagate inaccuracies in your models.
